```python
def c64convert(line: str) -> bytes:
    # for each character in line
    # convert to PETSCII
    res = bytearray()
    for c in line:
        if c == ' ':
            res.append(0x20)
        # penguin
        elif c == '§':
            res.append(0x5d)
        elif c == 'Ě':
            res.append(0x5e)
        elif c == 'Š':
            res.append(0x5f)
        elif c == 'Č':
            res.append(0x6d)
        elif c == 'Ř':
            res.append(0x6e)
        elif c == 'Ž':
            res.append(0x6f)
        elif c == 'Ý':
            res.append(0x7d)
        elif c == 'Á':
            res.append(0x7e)
        elif c == 'Í':
            res.append(0x7f)
        # sky
        elif c == 'ů':
            res.append(0x6b)  # Char index
        elif c == 'á':
            res.append(0x6c)
        elif c == 'č':
            res.append(0x7b)
        elif c == 'ď':
            res.append(0x7c)
        # Prachatice
        elif c == 'ú':
            res.append(0x69)
        elif c == 'é':
            res.append(0x6a)
        elif c == 'ě':
            res.append(0x79)
        elif c == 'š':
            res.append(0x7a)

        elif c.islower():
            res.append(ord(c) - ord('a') + 0x1)
        elif c.isupper():
            res.append(ord(c) - ord('A') + 0x41)
        elif c.isdigit():
            res.append(ord(c) - ord('0') + 0x30)
        elif c == '.':
            res.append(0x2e)
        elif c == ',':
            res.append(0x2c)
        elif c == '!':
            res.append(0x21)
        elif c == '?':
            res.append(0x3f)
        elif c == ':':
            res.append(0x3a)
        elif c == ';':
            res.append(0x3b)
        elif c == '"' or c == '“' or c == '”':
            res.append(0x22)
        elif c == "'" or c == '’':
            res.append(0x27)
        elif c == '(':
            res.append(0x29)
        elif c == ')':
            res.append(0x28)
        elif c == '-' or c == '—':
            res.append(0x2d)
        elif c == '+':
            res.append(0x2b)
        elif c == '/':
            res.append(0x2f)
        elif c == '*':
            res.append(0x2a)
        elif c == '=':
            res.append(0x3d)
        elif c == '<':
            res.append(0x3c)
        elif c == '>':
            res.append(0x3e)
        elif c == '~':
            res.append(0x63)
        elif c == '`':
            res.append(0x78)
        elif c == '%':
            res.append(0x25)
        elif c == '&':
            res.append(0x26)
        elif c == '#':
            res.append(0x23)
        else:
            print(f"Unknown character: {c}")
    return bytes(res)
```

#### Google search results static page

>Convert results-text.src to C64 screen codes

The load address is $5a00

```python
c64 = bytearray([0x00, 0x5a])  # set load address here

with open('assets/results-text.src', 'r') as f:
    for line in f:
        # if line is shorter than 40 chars complete it with spaces to 40 chars
        line = line.rstrip('\n')
        # split line into 40 char chunks
        line_chunks = [line[i:i+40] for i in range(0, len(line), 40)]
        for line in line_chunks:
            line = line.ljust(40, ' ')
            # convert to bytes and extend c64 byte array
            c64.extend(c64convert(line))

with open('../data/results-text.bin', 'wb') as f:
    f.write(c64)
```
